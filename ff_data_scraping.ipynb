```python
# First get all the unique players
def add_names_and_positions(names, positions, dfs, position):
    initial_names = []
    for df in dfs:
        initial_names.append(df['Unnamed: 0_level_0']['Player'].tolist())
    new_names = {x for l in initial_names for x in l}
    
    for name in new_names:
        index_period = name.rindex(".")
        names.append(name[0:index_period - 1])
        positions.append(position)

names = []
positions = []
add_names_and_positions(names, positions, [qb2020_df, qb2019_df, qb2018_df, qb2017_df, qb2016_df, qb2015_df, qb2014_df, qb2013_df, qb2012_df], "QB")
add_names_and_positions(names, positions, [rb2020_df, rb2019_df, rb2018_df, rb2017_df, rb2016_df, rb2015_df, rb2014_df, rb2013_df, rb2012_df], "RB")
add_names_and_positions(names, positions, [wr2020_df, wr2019_df, wr2018_df, wr2017_df, wr2016_df, wr2015_df, wr2014_df, wr2013_df, wr2012_df], "WR")
add_names_and_positions(names, positions, [te2020_df, te2019_df, te2018_df, te2017_df, te2016_df, te2015_df, te2014_df, te2013_df, te2012_df], "TE")
player_ids = range(len(names))
player_teams = [''] * len(names)
age = [-1] * len(names)

player_df = pd.DataFrame({
    'playerId': player_ids,
    'name': names,
    'nflTeamName': player_teams,
    'position': positions,
    'age': age
})
player_df.to_csv("./players.csv", index=False)
player_df
```

```python
def process_seasons(dfs, year, years, p_ids, passing_yards, passing_tds, interceptions, rushing_yards, rushing_touchdowns, receiving_yards, receiving_touchdowns, fumbles, receptions):
    for df in dfs:
        for index, row in df.iterrows():
            years.append(year)
            
            # Get the player id
            name = row['Unnamed: 0_level_0']['Player']
            index_period = name.rindex(".")
            name = name[0:index_period - 1]
            p_ids.append(player_ids[names.index(name)])

            # Pipe in the rest of the columns
            passing_yards.append(row['Passing']['Yds'])
            passing_tds.append(row['Passing']['TD'])
            interceptions.append(row['Passing']['Int']) 
            
            rushing_yards.append(row['Rushing']['Yds'])
            rushing_touchdowns.append(row['Rushing']['TD'])
            
            receiving_yards.append(row['Receiving']['Yds'])
            receiving_touchdowns.append(row['Receiving']['TD'])
            receptions.append(row['Receiving']['Rec'])
            
            fumbles.append(row['Fumbles']['FL'])


p_ids = []
passing_yards = []
passing_tds = []
interceptions = []
rushing_yards = []
rushing_touchdowns = []
receiving_yards = []
receiving_touchdowns = []
fumbles = []
receptions = []
years = []

dfs_2020 = [qb2020_df, rb2020_df, wr2020_df, te2020_df]
dfs_2019 = [qb2019_df, rb2019_df, wr2019_df, te2019_df,]
dfs_2018 = [qb2018_df, rb2018_df, wr2018_df, te2018_df]
dfs_2017 = [qb2017_df, rb2017_df, wr2017_df, te2017_df]
dfs_2016 = [qb2016_df, rb2016_df, wr2016_df, te2016_df]
dfs_2015 = [qb2015_df, rb2015_df, wr2015_df, te2015_df]
dfs_2014 = [qb2014_df, rb2014_df, wr2014_df, te2014_df]
dfs_2013 = [qb2013_df, rb2013_df, wr2013_df, te2013_df]
dfs_2012 = [qb2012_df, rb2012_df, wr2012_df, te2012_df]

process_seasons(dfs_2020, 2020, years, p_ids, passing_yards, passing_tds, interceptions, rushing_yards, rushing_touchdowns, receiving_yards, receiving_touchdowns, fumbles, receptions)
process_seasons(dfs_2019, 2019, years, p_ids, passing_yards, passing_tds, interceptions, rushing_yards, rushing_touchdowns, receiving_yards, receiving_touchdowns, fumbles, receptions)
process_seasons(dfs_2018, 2018, years, p_ids, passing_yards, passing_tds, interceptions, rushing_yards, rushing_touchdowns, receiving_yards, receiving_touchdowns, fumbles, receptions)
process_seasons(dfs_2017, 2017, years, p_ids, passing_yards, passing_tds, interceptions, rushing_yards, rushing_touchdowns, receiving_yards, receiving_touchdowns, fumbles, receptions)
process_seasons(dfs_2016, 2016, years, p_ids, passing_yards, passing_tds, interceptions, rushing_yards, rushing_touchdowns, receiving_yards, receiving_touchdowns, fumbles, receptions)
process_seasons(dfs_2015, 2015, years, p_ids, passing_yards, passing_tds, interceptions, rushing_yards, rushing_touchdowns, receiving_yards, receiving_touchdowns, fumbles, receptions)
process_seasons(dfs_2014, 2014, years, p_ids, passing_yards, passing_tds, interceptions, rushing_yards, rushing_touchdowns, receiving_yards, receiving_touchdowns, fumbles, receptions)
process_seasons(dfs_2013, 2013, years, p_ids, passing_yards, passing_tds, interceptions, rushing_yards, rushing_touchdowns, receiving_yards, receiving_touchdowns, fumbles, receptions)
process_seasons(dfs_2012, 2012, years, p_ids, passing_yards, passing_tds, interceptions, rushing_yards, rushing_touchdowns, receiving_yards, receiving_touchdowns, fumbles, receptions)


season_df = pd.DataFrame({
    'seasonId': range(len(p_ids)),
    'playerId': p_ids,
    'passingYards': passing_yards,
    'passingTouchdowns': passing_tds,
    'interceptions': interceptions,
    'rushingYards': rushing_yards,
    'fumbles': fumbles,
    'receptions': receptions,
    'receivingTouchdowns': receiving_touchdowns,
    'receivingYards': receiving_yards,
    'rushingTouchdowns': rushing_touchdowns,
    'year': years
})
season_df.to_csv("./seasons.csv", index=False)
```
